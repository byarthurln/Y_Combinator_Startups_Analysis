# Modification of team size values

```python
# Handle missing or non-finite values in the 'team_size' column
df['team_size'] = df['team_size'].fillna(0)  # Replace missing values with 0 or another default value
# Check if values are finite before converting to integers
df['team_size'] = df['team_size'].apply(lambda x: int(x) if pd.notna(x) and not np.isinf(x) else 0)
# Save the updated data
df.to_csv('/content/sample_data/2024-05-11-yc-companies.csv', index=False)
df.head()
```

```python
# Check for missing or empty values in the 'year_founded' column
nan_or_empty_rows = df[df['year_founded'].isna() | (df['year_founded'].astype(str).str.strip() == '')]

# Display the number of rows where the 'year_founded' column is NaN or empty
print("\nNumber of rows where the 'year_founded' column is NaN or empty:")
print(len(nan_or_empty_rows))
```

#Hypothesis: Replace null values in "year_founded" with the year corresponding to the startups' batches

```python
# Function to extract the year from the batch
def extract_year_from_batch(batch):
    if pd.notna(batch):
        if batch.startswith('W') or batch.startswith('S'):
            if len(batch) == 3 and batch[1:].isdigit():
                year_str = batch[1:]  # Remove the first character (W or S)
                return float(f"20{year_str}")
        elif batch.startswith('IK'):
            if len(batch) == 4 and batch[2:].isdigit():
                year_str = batch[2:]  # Remove the first two characters (IK)
                return float(f"20{year_str}")
    return None

# Apply the function to the 'year_founded' column to fill missing or null values
def update_year_founded(row):
    if pd.isna(row['year_founded']) or row['year_founded'] == '':
        year_founded = extract_year_from_batch(row['batch'])
        if year_founded is None:
            print(f"No valid batch found for {row['company_name']} with batch: {row['batch']}")
        return year_founded
    return row['year_founded']

df['year_founded'] = df.apply(update_year_founded, axis=1)
# Save the updated data
df.to_csv('/content/2024-05-11-yc-companies.csv', index=False)
df.head()
```

```python
# Check for missing or empty values in the 'year_founded' column
nan_or_empty_rows = df[df['year_founded'].isna() | (df['year_founded'].astype(str).str.strip() == '')]

# Display the number of rows where the 'year_founded' column is NaN or empty
print("\nNumber of rows where the 'year_founded' column is NaN or empty:")
print(len(nan_or_empty_rows))
```

```python
# Convert the values in the 'year_founded' column to integers
df['year_founded'] = df['year_founded'].astype(int)
df.head()
```

```python
# Check if columns exist before dropping them
columns_to_drop = ['cb_url', 'linkedin_url']
for column in columns_to_drop:
    if column in df.columns:
        df = df.drop(columns=[column])
df.head()
```

#Split the 'tags' column into multiple separate columns

```python
# Check if the 'tags' column exists
if 'tags' in df.columns:
    # Remove brackets and transform the 'tags' column into a Python list
    df['tags'] = df['tags'].apply(lambda x: ast.literal_eval(x))

    # Create new columns tag_1, tag_2, etc., based on the content of 'tags'
    max_tags = df['tags'].apply(len).max()
    for i in range(1, max_tags + 1):
        df[f'tag_{i}'] = df['tags'].apply(lambda x: x[i-1] if i-1 < len(x) else None)

    # Remove the 'tags' column
    df = df.drop(columns=['tags'])

# Add the 'quantity' column with the value 1 in each cell
df['quantity'] = 1
# Save the updated data
df.to_csv('/content/2024-05-11-yc-companies_NEW.csv', index=False)
df.head()
```
